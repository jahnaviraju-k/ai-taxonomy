```python
import pandas as pd
import plotly.express as px

# Load the Excel file
file_path = 'Mapped HITL Interaction to Algorithm Data.xlsx'
excel_data = pd.ExcelFile('C:/Users/HP/Downloads/Mapped HITL Interaction to Algorithm Data.xlsx')
# Parse the sheet "Algorithms"
data = excel_data.parse('Algorithms')

# Prepare data by grouping algorithms by Interaction Types
filtered_data = data[['Algorithms', 'Interaction Types']].dropna()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Count the number of algorithms per interaction type
interaction_counts = filtered_data['Interaction Types'].value_counts()

# Standardize the 'Interaction Types' column by converting to lowercase and removing extra spaces
data['Interaction Types'] = data['Interaction Types'].str.strip().str.capitalize()

# Count the occurrences of each interaction type
interaction_counts = data['Interaction Types'].value_counts()

# Plotting the distribution of interaction types
plt.figure(figsize=(8, 6))
interaction_counts.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Distribution of Interaction Types')
plt.xlabel('Interaction Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(interaction_counts, labels=interaction_counts.index, autopct='%1.1f%%', startangle=140, colors=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Proportion of Interaction Types')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(interaction_counts, labels=interaction_counts.index, autopct='%1.1f%%', startangle=140, colors=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'], wedgeprops=dict(width=0.3))
plt.title('Proportion of Interaction Types')
plt.show()
```
